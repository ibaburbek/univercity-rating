```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
places2=[]
faculties2=[]
names2 = []
regions2 = []
career = []
expert = []
stats = []
total = []
salary = []
employment = []
duration = []




url = 'https://atameken.kz/ru/university_ratings?page=1'

page = requests.get(url, verify=False)

#print(page.status_code)
soup = BeautifulSoup(page.text, 'html.parser')

table = soup.find('div', class_='universities')






column_names = []
column_names.append('Место')
column_names.append('Факультет')
column_names.append('Название')
column_names.append('Регион')
for i in range(7):
    column_names.append(table.find_all('tr')[i].text.split('\n')[1])




for i in range(1,176):



    url = 'https://atameken.kz/ru/university_ratings?page=' + str(i)
    page = requests.get(url, verify=False)
    soup = BeautifulSoup(page.text, 'html.parser')
    table = soup.find('div', class_='universities')
    places = table.find_all('div', class_= 'university__place')
    for place in places:
        places2.append(place.text)
    faculties = table.find_all('div', class_='university__faculty')
    for i in range(len(faculties)//3):
        k=3*i
        faculties2.append(faculties[k].text + '---' + faculties[k+1].text + '---' + faculties[k+2].text)

    names = table.find_all('h3', class_='university__name')
    for name in names:
        names2.append(name.text)

    regions = table.find_all('div', class_='university__region')
    for region in regions:
        regions2.append(region.text)

    for j in range(12):
        x=table.find_all('tr')
        career.append(x[7*j].text.split('\n')[2])
        expert.append(x[7*j+1].text.split('\n')[2])
        stats.append(x[7*j+2].text.split('\n')[2])
        total.append(x[7*j+3].text.split('\n')[2])
        salary.append(x[7*j+4].text.split('\n')[2])
        employment.append(x[7*j+5].text.split('\n')[2])
        duration.append(x[7*j+6].text.split('\n')[2])
```

```python
df=pd.DataFrame(list(zip(places2, faculties2, names2, regions2, career,expert,stats,total, salary, employment,duration)), columns=column_names)
df.index +=1
df
```

```python
df.to_excel('university_ratings.xlsx')
```
